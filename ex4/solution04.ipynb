```python
import numpy as np
import matplotlib.pyplot as plt
```

**4.1: Inverse CDF and Random Number Generation**

```python
def p(x, m=1, b=2):
    return 1 / (2 * b) * np.exp(-(np.abs(x - m) / b))

def f_inverse(y, m=1, b=2):
    return np.where(y >= 0.5, m - b * np.log(2 - 2 * y), m + b * np.log(2 * y))

samples = np.asarray([f_inverse(y) for y in np.random.random(500)])
 
x = np.linspace(samples.min(), samples.max(), 700)
pdf = p(x)

plt.hist(samples, bins=70, density=True)
plt.plot(x, pdf)
plt.xlabel('samples')
plt.ylabel('pdf')
plt.show()
```
